These visualizations are unique but not interactive. In order to create an interactive graph with D3.js, we will have to create a JSON file from the provided GML.

```python
g = nx.read_gml('dolphins.gml')
#Create a dictionary in a node-link format that is suitable for JSON serialization
d = json_graph.node_link_data(g)
#Create a new file dolphins.json, and write to it:
with open('dolphins.json', 'w') as fp:
    json.dump(d, fp)
```

```python
#Print the new JSON data to view format and understand the data:
d
```

<b>Understanding the data format:</b><br>
The nodes are labeled with the ID of the dolphin number and name. They start at 0. <br>
The links contain the source (Dolphin number) and the target (Another dolphin number)<br>
  {'id': 55, 'label': 'Vau'},<br>
  Dolphin 'Vau' only has two encounters on the visualization,<br> https://raw.githubusercontent.com/jeremyshantz/dolphin-social-network/master/dolphin.png<br>
  'Haecksel' and 'Trigger'<br>
{'source': 15, 'target': 55}, - Haecksel is ID 15<br>
{'source': 51, 'target': 55}, - Trigger is ID 51<br>
The encounters are not repeated more than one time.
